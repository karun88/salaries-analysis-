```python
import pandas as pd
import matplotlib.pyplot as plt  # Corrected import statement
import seaborn as sns

data = pd.read_csv(r"C:\Users\karun sai.D\OneDrive\Desktop\Salaries\salaries.csv")
```

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
# Salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Salary'], kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# Average salary by job
plt.figure(figsize=(12, 8))
sns.barplot(x='Job', y='Salary', data=data.groupby('Job').agg({'Salary':'mean'}).reset_index())
plt.title('Average Salary by Job')
plt.xlabel('Job')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation matrix
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Salary vs Age scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=data)
plt.title('Salary vs Age')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
